```python
from langgraph.graph import START, END, StateGraph
from typing import TypedDict
```

```python
class BatsmanState(TypedDict):
    runs: int = 10
    balls: int = 200
    fours: int
    sixes: int

    sr: float
    bpb: float
    boundary_percent: float
    summary: str
```

```python
def strike_rate(state: BatsmanState):
    sr = (state['runs']/state['balls'])*100
    return {'sr': sr}

def balls_per_boundary(state: BatsmanState):
    bpb = state['balls']/ (state['fours'] + state['sixes'])
    return {"bpb": bpb}

def boundary_percent(state: BatsmanState) -> BatsmanState:
    bp = (((state['fours'] * 4) + (state['sixes'] * 6))/state['runs'])*100
    return {'boundary_percent': bp}

def summary(state: BatsmanState) -> BatsmanState:
    summary = f"""
        Strike Rate - {state['sr']} \n
        Balls per boundary - {state['bpb']} \n
        Boundary percent - {state['boundary_percent']}
        """
    return {'summary': summary}
```

```python
graph = StateGraph(BatsmanState)

graph.add_node('sr', strike_rate)
graph.add_node('bpb', balls_per_boundary)
graph.add_node('bp', boundary_percent)
graph.add_node('sm', summary)

graph.add_edge(START, 'sr')
graph.add_edge(START, 'bpb')
graph.add_edge(START, 'bp')

graph.add_edge('sr', 'sm')
graph.add_edge('bpb', 'sm')
graph.add_edge('bp', 'sm')

graph.add_edge('sm', END)
```

```python
workflow = graph.compile()
```
